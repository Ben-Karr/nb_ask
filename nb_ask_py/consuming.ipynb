```python
import httpx
from pprint import pprint
import json

from lisette import *
import os
from dotenv import load_dotenv
```

```python
def get_cells():
    ret = httpx.get("http://localhost:3000/")
    cell_content = ret.json()
    return cell_content
```

```python
load_dotenv()
endpoint = os.environ["OPENAI_ENDPOINT"]
key = os.environ["OPENIA_API_KEY"]
```

```python
cells_content = get_cells()
```

Test without sp:

```python
chat = Chat(
    model="azure_ai/gpt-4-turbo-1106",
    api_base=endpoint,
    api_key=key,
)
```

```python
pr = "Can you describe what im doing in this notebook?"
chat(f"active notebook content: {json.dumps(cells_content)} | prompt: {pr}")
```

Test with sp

```python
sp = "Allways be helpful and suggestive. Make an efford to always give the shortest answer possible without losing correctness."
chat = Chat(
    model="azure_ai/gpt-4-turbo-1106",
    api_base=endpoint,
    api_key=key,
    sp=sp
)
chat(f"active notebook content: {json.dumps(cells_content)} | prompt: {pr}")
```

```python
sp = """Allways be helpful and suggestive. Make an efford to always give the shortest answer possible without losing correctness. In addition to the actual promt you are given the content of the jupyter notebook you are currently in - it's inbetween
 `active notebook content: ` and ` |`. Each item of the json serialized list represents one cell consisting of `index` which gives you the execution order; `kind` so if its a code or a markdown cell; `source` is the user input to the cell; `output`
 is the result of the cell execution."""
 
chat = Chat(
    model="azure_ai/gpt-4-turbo-1106",
    api_base=endpoint,
    api_key=key,
    sp=sp
)
chat(f"active notebook content: {json.dumps(cells_content)} | prompt: {pr}")
```
